## Tipos de datos

* Booleanos (`bool`): <hfill/> `True` y `False`.
* Enteros (`int`): <hfill/> `0`, `1`, `2`, `-1`, `-2`, etc.
* *Reales* (`float`): <hfill/> `1.2`, `3.4e5`, `float('inf')` ($+\infty$), `float('nan')`, etc.
* Números complejos (`complex`): <hfill/> `complex(2,3)`, `2+3j`, etc.
* Cadenas de texto (`str`): <hfill/> `''`, `"una"` `'otra'`, `'''varias líneas'''`, etc.
* Tuplas (`tuple`): <hfill/> `()`, `(1, 2)`, `(1, 'uno')`, `(1.6, 2, 3.6)`, etc.
* Listas (`list`): <hfill/> `[]`, `[1, 2]`, `[1, 'uno']`, `[1.6, 2, 3.6]`, etc.
* Ausencia de valor (`NoneType`): <hfill/> `None`

### Secuencias

Las cadenas de texto, las tuplas y las listas se pueden englobar en la categoría de *secuencias*. Dado un objeto `x` de cualquiera de estos tres tipos, es posible
* calcular su longitud `len(x)`.
* acceder al elemento $n$-ésimo con `x[n]` (el primero es el 0).
* acceder al $n$-ésimo por el final con `x[-n]` (el último es el -1).
* obtener la subsecuencia entre los índices $n$ (inclusive) y $m$ (exclusive) con `x[n:m]`. Si $n = 0$ o $m =$ `len(x)` se pueden omitir (i.e., `x[:m]`, `x[n:]`, x`[:]`). También se puede añadir un tercer número `x[n:m:k]` para saltar de $k$ en $k$ índices.
* comprobar si contiene un elemento `e` con `e in x`.
* obtener la concatenación con otra secuencia `y` del mismo tipo con `x + y`.
* obtener la secuencia repetida $k$ veces con `x * k`.
* encontrar el índice de la primera aparición de `e` con `x.index(e)`.
* calcular el número de apariciones de `e` con `x.count(e)`.
* comprobar si es igual a otra secuencia `y` con `x == y`.
* compararla lexicográficamente con otra secuencia `y` con `x < y`, `x >= y`, etc.

La definición oficial de secuencia en Python es menos específica e incluye únicamente los dos primeros puntos.

### Tuplas

Secuencias inmutables de cualquier combinación de tipos de datos (es decir, tuplas en el sentido habitual, elementos de un producto cartesiano $A_1 \times \cdots \times A_n$). Se escriben enumerando sus elementos separados por comas (`,`) con paréntesis opcionales. La tupla de 0 elementos se escribe `()`.

Son útiles para devolver múltiples valores en las funciones. La asignación múltiple permite descomponerlas y asignar sus elementos a variables. Por ejemplo, `x, y = par` donde `par = (1, 2)`.

### Listas

Secuencias mutables de cualquier combinación de tipos de datos (es decir, tuplas mutables), aunque el tipo de todos sus elementos suele ser el mismo. Se escriben también enumerando sus elementos separados por comas (`,`), pero entre corchetes (`[]`). La lista vacía se escribe `[]`. A diferencia de las tuplas, es posible
* reemplazar el elemento $k$-ésimo por `e` con `x[k] = e`.
* añadir un nuevo elemento `e` al final con `x.append(e)`.
* añadir todos los elementos de otra lista `y` al final con `x.extend(y)` o `x += y`.
* insertar un elemento `e` en la posición $k$ con `x.insert(k, e)`.
* eliminar el elemento en la posición $k$ con `x.pop(k)` o `del x[k]`.
* eliminar la primera aparición de `e` con `x.remove(e)`

También se puede ordenar la lista de menor a mayor con `x.sort()` e invertir su orden con `x.reverse()`.

### Cadenas de texto

Son secuencias inmutables de caracteres (letras, números, símbolos, espacios, etc). Se escriben como texto entre comillas simples (`'hola'`) o dobles (`"hola"`). Las cadenas pueden extenderse varias líneas usando tres comillas (`'''` o `"""`) en lugar de una. La barra invertida (`\`) seguida de ciertos caracteres permite introducir caracteres especiales: `\n` para el salto de línea, `\t` para la tabulación, `\\` para la propia barra invertida, `\'` o `\"` para escribir la comilla en lugar de terminar la cadena, entre otros.

Además de los métodos comunes a la secuencias, se les pueden aplicar [otros métodos](https://docs.python.org/3/library/stdtypes.html#string-methods) para
* buscar una subcadena `s` dentro de `x` y obtener el índice donde empieza su primera aparición con `x.find(s)`. Si no se encuentra, el resultado es `-1`. El método `index` y el operador `in` de las secuencias admiten también cadenas como primer argumento, no solo con caracteres sueltos. Por ejemplo, `'th' in 'Python'`.
* pasar una palabra a mayúsculas o minúsculas con `x.lower()` o `x.upper()`.
* reemplazar una subcadena `s` por otra `r`con `x.replace(s, r)`.
* separarla por cada aparición de la subcadena `sep` en una lista de cadenas con `x.split(sep)`. Con `x.split()` la cadena se separa por palabras. Por ejemplo, `'enero de 2023'.split()` da lugar a `['enero', 'de', '2023']`.
* juntar una secuencia de cadenas `xs` utilizando `x` como separador con `x.join(xs)`. Por ejemplo, `','.join(['1', '2', '3'])` resulta `'1,2,3'`.

Y sobre caracteres (es decir, cadenas de tamaño uno) las siguientes funciones son aplicables:
* obtener su codificación numérica en [Unicode](https://es.wikipedia.org/wiki/Unicode) con `ord(x)`. Por ejemplo, `ord('a')` es  97.
* obtener un carácter a partir de su codificación numérica con `chr(n)`. Por ejemplo, `chr(123)` es `'{'`.

Si la cadena se precede de la letra `f` como en `f'1 + 2 = {1 + 2}'`, el texto entre cada par de corchetes (`{}`) se evaluará como una expresión de Python y sustituirá por el resultado. Por ejemplo, en el caso anterior quedará `'1 + 2 = 3'`. Los corchetes literales se escriben duplicando el símbolo (`{{` o `}}`). Más información en el [manual](https://docs.python.org/3/library/string.html#format-string-syntax) de Python.

## Instrucciones

Las instrucciones del lenguaje Python se pueden clasificar en *expresiones* y *sentencias* (o *instrucciones de control*). Las primeras producen un valor, que se puede asignar a una variable o pasar como argumento a una función. Se forman con
* Constantes: <hfill/> `1`, `2.3`, `"sol"`, `math.pi`, etc.
* Variables.
* Operadores aritméticos: `+`, `-`, `*`, `/`, `//` para la división entera, `%` para el resto de la división entera, `**` para la potencia. Su significado varía según el tipo de datos.
* Operadores lógicos: <hfill/> `and`, `or` y `not`.
* Comparadores: <hfill/> `==`, `!=` `<`, `<=`, `>`, `>=`, etc.
* Llamadas a función: <hfill/> `f(e1, ..., en)`.
* Conversión de tipos: <hfill/> `int('1')`, `float('2.3')`, `str(3.4)`, etc.
* Acceso a los miembros de un objeto: <hfill/> `(2+3j).real`, `'hola'.upper`, etc.
* Expresión condicional: <hfill/> `'días' if hora < 12 else 'tardes'`.

Los comparadores en Python se pueden concatenar, es decir, se puede escribir `0 <= k < 10` en lugar de `0 <= k and k < 10`.

Las sentencias no devuelven ningún valor, pero permiten modificar el estado del programa y tomar decisiones para ejecutar unas partes u otras del código. Entre ellas están

* Asignación: simple `x = e` o simultánea `x, y, z = e, f, g`.
* Eliminación en una lista: `del x[k]`.
* Condicionales.
* Bucles.

En las siguientes secciones se presentan algunas más relacionadas con bucles (`break`, `continue`) y funciones (`return e`).

### Condicionales

Una instrucción condicional ejecuta un único bloque de código de entre un conjunto de ellos, evaluando las expresiones booleanas $e_k$ que los preceden sucesivamente hasta encontrar una que se evalúe a `True`. El bloque correspondiente a esa expresión será el único que se ejecute. Si ninguna condición se satisface, no se ejecutará ningún bloque o se ejecutará el bloque `else`, en caso de haberlo.

```python
if e1:
    # <bloque de código si e1>
elif e2:
    # <bloque de código si ¬ e1 y e2>
elif e3:
    # <bloque de código si ¬ e1 y ¬ e2 y e3>
# [...]
else:
    # <bloque de código si ninguna de las anteriores>
```

Cada bloque ha de estar identado (es decir, más a la derecha) respecto al `if`. Se pueden usar espacios en la cantidad deseada o tabulaciones para identar el código, pero hay que ser consistente. El código a ejecutar tras el condicional ha de estar al mismo nivel que el `if`.

## Bucles

Los bucles ejecutan repetidamente una parte del programa hasta que se satisfaga una condición. El bucle más sencillo es el `while`
```python
while e:  # e es la condición
    # <bloque de código que se repetirá mientras se cumpla e>
```
Habitualmente `e` dependerá de las variables del cuerpo del bucle y llegará a valer `False` en algún momento para que el programa termine. Es típico hacer recorridos con índices como este
```python
k = valor_inicial

while k < tope:
    # <código que hace algo>
    k += paso
```
que se pueden simplificar con el bucle `for` y `range`
```python
for k in range(valor_inicial, tope, paso):
    # <código que hace algo>
```
donde `paso` se puede omitir si es 1 y `valor_inicial` si es 0. En general, la expresión a la derecha de `in` puede ser cualquier objeto *iterable* (cadenas, tuplas, listas, etc).

Es frecuente usar bucles para hacer búsquedas, que conviene interrumpir en cuanto se haya encontrado lo buscado. En el bucle `while` esto se puede comprobar en la condición, pero no así en `for`. Tanto en `while` como en `for` también es posible interrumpir el bucle utilizando la sentencia `break`. La sentencia `continue` permite pasar a la siguiente iteración sin terminar de ejecutar el resto del cuerpo del bucle.

## Funciones

Sirven para organizar el código en componentes reutilizables y abstraer detalles de implementación. Se declaran como:
```python
def nombre_función(nombre_argumento_1, ..., nombre_argumento_n):
    # <cuerpo (identado)>
```
Las funciones pueden devolver un valor al llamante, modificar los objetos mutables que hayan recibido como parámetro y tener efectos secundarios como imprimir texto por pantalla. La instrucción `return e` termina la ejecución de la función (el código tras ella no se ejecutará) devolviendo el valor de la expresión `e`. Si una función termina sin pasar por una instrucción `return` se devolverá `None`.

Se llama a una función escribiendo `nombre_función(e1, ..., en)` donde `ek` son expresiones. También es posible escribir el nombre completo de los parámetros formales
```python
nombre_función(e1, e2, nombre_argumento_3=e3, ..., nombre_argumento_n=en)
```
aunque no puede pasarse un argumento sin nombre después de uno con nombre. En la definición de la función, se pueden fijar valores por defecto `nombre_arg=valor`, con lo que no será necesario proporcionar esos argumentos en la llamada salvo que se les quiera dar otro valor.

Python proporciona algunas funciones predefinidas de uso común:

* [`print`](https://docs.python.org/3/library/functions.html#print) para imprimir texto por el terminal. Por defecto, escribe la representación textual de todos sus argumentos separados por un espacio y termina con un salto de línea.
* [`input`](https://docs.python.org/3/library/functions.html#input) para leer una línea desde el terminal. Recibe como argumento opcional un texto a modo de entradilla que se imprimirá para indicar al usuario lo que se espera de él.
* [`round`](https://docs.python.org/3/library/functions.html#round) para redondear un número. Por ejemplo, `round(1 / 3, 2)` es `0.33`.

## Esquemas algoritmos

Algunos esquemas algorítmicos que han aparecido en el primer cuatrimestre.

### Búsqueda exhaustiva

Se busca exhaustivamente en un conjunto de objetos aquel que cumple una determinada propiedad.

```python
def es_primo(n: int) -> bool:
    """Comprueba si un número es primo"""
    
    # Busca un divisor entre todos los números hasta √n
    for d in range(2, int(n ** 0.5) + 1):
        if n % d == 0:
            return False
    
    # No tiene ningún divisor, luego es primo
    return True
```

### Búsqueda por bisección

Se busca una solución aproximada a un problema dividiendo sucesivamente el espacio de soluciones en dos mitades y descartando una de ellas.

```python
def sqrt(n: float, eps : float = 0.01) -> float:
    """Busca la raíz cuadrada de un número dado admitiendo un cierto error"""
    
    # Si el número es negativo no hay raíz cuadrada (real)
    if n < 0.0:
        return None
    
    # La raíz cuadrada está comprendida entre estos dos números
    menor, mayor = 0.0, n / 2 + 1
    
    # Prueba con el número medio
    medio = (menor + mayor) / 2
    
    while abs(medio ** 2 - n) >= eps:
        if medio ** 2 < n:
            menor = medio
        else:
            mayor = medio
        
        medio = (menor + mayor) / 2
    
    return medio
```
